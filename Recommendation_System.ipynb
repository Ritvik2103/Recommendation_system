### DERIVING UNIQUE DIRECTORS

```python
dire=[]
for i in df['director']:
    dire.extend(i.split(','))
dire= list(set(dire))
```

```python
len(dire)
```

### CREATE ONE HOT ENCODED

```python
one_hot=[]
for i in df['director']:
    enc=[0]*1159
    for j in i.split(','):
        enc[dire.index(j)]=1
    one_hot.append(enc)
```

```python
len(one_hot)
```

```python
one_hot= np.array(one_hot)
```

```python
one_hot.shape
```

### VERIFYING ONE HOT

```python
for i in range(len(df['director'])):
    x= list(one_hot[i]).count(1)
    y= len(df['director'][i].split(','))
    if(x !=y):
        print('Not Accurate')
        break
    else:
        flag=1
if(flag):
    print('Accurate One Hot')
```

### Adding One hot to dataset

```python
df
```

```python
for i in range(len(df['director'])):
    df[dire[i]]= one_hot[: , i]
    
```

#### RECOMMENDING BASED ON THE SIMILAR DIRECTORS

```python
def func(list):
    sm=0
    for i in list:
        sm+= abs(i)
        
    return sm
```

```python
user= input('Enter Name: ')
for i in df.values:
    if(user.lower()== i[0].lower()):
        rec= np.array(i[7:])
lst=[]
for i in df.values:
    diff=func(list(np.array(i[7:])-rec))
    lst.append([i[0], diff])
df_= pd.DataFrame(lst, columns=['Movies', 'Difference'])
print('Recommended Movies/Shows: ')
for i in df_.sort_values(by='Difference').head(10).iloc[:, 0].values:
    print(i)
```

### NAME BASED RECOMMENDATIONS

```python
df= pd.read_csv('imdb.csv')
df= df[df['genre'] != 'Nan']
```

```python
df
```

```python
name=[]
for i in df['name']:
    i=i.replace(':', '')
    i= i.replace('-', '')
    if(len(i.split(' '))<=3):
        name.append(' '.join(i.split(' ')[:len(i.split(' '))//2 +1]))
    else:
        name.append(' '.join(i.split(' ')[: len(i.split(' '))//2 ]))
name= list(set(name))
print(len(name))
```

```python
one_hot=[]
for i in df['name']:
    i= i.replace(':', '')
    i= i.replace('-', '')
    enc=[0]*1169
    if(len(i.split(' '))<=3):
        enc[name.index(' '.join(i.split(' ')[:len(i.split(' '))//2 +1]))]=1
    else:
        enc[name.index(' '.join(i.split(' ')[: len(i.split(' '))//2 ]))]=1
    one_hot.append(enc)
    
print(len(one_hot))
    
    
```

### Verifying one hot

```python
for i in range(len(df['name'])):
    if(one_hot[i].count(1) != 1):
        print('Not valid')
        break
    else:
        flag=1
if(flag):
    print('Valid One Hot Encoded!!')
```

### adding encoded columns to data set

```python
one_hot= np.array(one_hot)
print(one_hot.shape)
```

```python
for i in (range(len(df['name']))):
    df[name[i]]=one_hot[: ,i]
```
